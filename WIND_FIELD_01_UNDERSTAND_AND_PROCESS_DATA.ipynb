### 3. Find the derived Variables

```python
# 1. Wind Speed
wind_speed = np.sqrt(u10**2 + v10**2)
```

```python
# 2. Relative Humidity
# Function to calculate vapor pressure and saturation vapor pressure
def calc_vapor_pressure(d2m):
    """Calculate actual vapor pressure (e) in hPa."""
    d2m_C = d2m - 273.15  # Convert from Kelvin to Celsius
    return 6.112 * np.exp((17.67 * d2m_C) / (d2m_C + 243.5))

def calc_saturation_vapor_pressure(t2m):
    """Calculate saturation vapor pressure (es) in hPa."""
    t2m_C = t2m - 273.15  # Convert from Kelvin to Celsius
    return 6.112 * np.exp((17.67 * t2m_C) / (t2m_C + 243.5))

def calc_relative_humidity(d2m, t2m):
    """Calculate relative humidity (RH) in percentage."""
    e = calc_vapor_pressure(d2m)  # Actual vapor pressure
    es = calc_saturation_vapor_pressure(t2m)  # Saturation vapor pressure
    return (e / es) * 100

relative_humidity = calc_relative_humidity(d2m, t2m)
```

```python
print(relative_humidity)
```

```python
# Function to calculate pressure gradient magnitude
def calc_pressure_gradient_magnitude(msl, lat, lon):
    """
    Calculate the magnitude of the surface pressure gradient in Pa/m.
    msl: 3D array (time, lat, lon) of mean sea-level pressure (Pa)
    lat: 1D array of latitudes
    lon: 1D array of longitudes
    """
    # Convert lat and lon to radians
    lat_rad = np.radians(lat)
    lon_rad = np.radians(lon)

    # Compute distances in meters between grid points
    dlat = np.gradient(lat_rad) * 6371000  # Earth's radius in meters
    # Convert lat_rad.mean() to a NumPy array and then to a float
    # to ensure it's a scalar
    dlon = np.gradient(lon_rad) * 6371000 * np.cos(lat_rad.mean().item())

    # Initialize gradient magnitude array
    gradient_magnitude = np.zeros_like(msl)

    # Loop over time dimension
    for t in range(msl.shape[0]):
        # Compute pressure gradients along lat and lon
        dp_dy = np.gradient(msl[t, :, :], axis=0) / dlat[:, None]
        dp_dx = np.gradient(msl[t, :, :], axis=1) / dlon[None, :]

        # Compute gradient magnitude
        gradient_magnitude[t, :, :] = np.sqrt(dp_dx**2 + dp_dy**2)

    return gradient_magnitude

# Calculate pressure gradient magnitude
pressure_gradient_magnitude = calc_pressure_gradient_magnitude(msl, lat, lon)
```

## 3. Merging the datasets

```python
# Rename variables to avoid conflicts
df1 = df1.rename_vars({'u10': 'u10', 'v10': 'v10'})
df2 = df2.rename_vars({'blh': 'blh'})
df3 = df3.rename_vars({'cape': 'cape'})
df4 = df4.rename_vars({'d2m': 'd2m'})
df5 = df5.rename_vars({'fsr': 'fsr'})
df6 = df6.rename_vars({'z': 'geo'})  # Rename 'z' to 'geo'
df7 = df7.rename_vars({'msl': 'msl'})
df8 = df8.rename_vars({'t2m': 't2m'})

# Merge all datasets
merged_ds = xr.merge([df1, df2, df3, df4, df5, df6, df7, df8])

# Check the merged dataset
print(merged_ds)
```

### 3.1 Assign the derived variables also

```python
# add wind magnitude
merged_ds['wind_magnitude'] = wind_speed

# Check the updated dataset
print(merged_ds)
```

```python
#add others also
merged_ds['RH'] = relative_humidity
merged_ds['spg'] = (('valid_time', 'latitude', 'longitude'), pressure_gradient_magnitude)
```
